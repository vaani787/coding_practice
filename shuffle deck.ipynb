# Python program to shuffle a deck of cards

```python
#importing modules
import itertools, random
```

```python
#make a deck of cards
deck=list(itertools.product(range(1,14),['Spade','Heart','Diamond','Club']))
```

```python
#shuffle the deck
random.shuffle(deck)
```

```python
#deck after shuffling
deck
```

```python
#draw five cards
for i in range(5):
   print("You got:{} of {}".format(deck[i][0], deck[i][1]))
```

```python
print("You got:")
for i in range(5):
   print(deck[i][0], "of", deck[i][1])
```

```python
# shuffle deck three times and draw two random card each time:
for i in range(3):
    random.shuffle(deck)
    for j in range(2):
        x=random.randint(1,52)     #randint(self, a, b)   #Return random integer in range [a, b], including both end points.
        print("You got:{} of {}".format(deck[x][0], deck[x][1]))
        
        
```
